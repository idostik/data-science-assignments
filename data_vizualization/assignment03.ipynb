A lot of the attributes have weird outliers, f.e. *AvgHouseholdSize*. We will remove them from the training dataset of the model.

Because some of the attributes are quite diverse, we will set the boundaries for removing outliers higher.

```python
def outliers(df, colname, remove=True):

    # Calculating quartiles
    Q1 = np.percentile(df[colname], 25,
                    method = 'midpoint') # first quartile
    
    Q3 = np.percentile(df[colname], 75,
                    method = 'midpoint') # third quartile
    IQR = Q3 - Q1
    print(IQR)
    
    # Upper bound
    upper = np.where(df[colname] >= (Q3+5*IQR))
    # Lower bound
    lower = np.where(df[colname] <= (Q1-5*IQR))
    
    print(f"Old row count before {colname}: ", df.shape[0])
    
    if remove:
        df.drop(upper[0], inplace=True, axis=0)
        df.drop(lower[0], inplace=True, axis=0)
        df.reset_index(drop=True, inplace=True)
        
        print(f"New row count after {colname}: ", df.shape[0])
    else:
        print("Rows that would be removed:", upper[0].shape[0] + lower[0].shape[0])

for column in train_data.columns:
    outliers(train_data, column)

print("Final row count: ", train_data.shape[0])
```

#### Model training - Ridge Regression

We will train a ridge regression model. In order to find the best hyperparameters, we will use ParameterGrid. The hyperparameters we will be tuning are *alpha* and *solver*.

*Alhpa* determines how much will the weights in ridge regression be penalized. The greater the *alpha*, the smaller weights will be chosen by the model. For *alpha* = 0, ridge regression is equal to classical linear regression.

The other hyperparameter decides which method will be used to solve the equation that will give us the smallest error in ridge regression. It may affect how stable the computations will be when given different results.

As our loss function we will use RMSE, which penalizes big deviations but opposed to MSE it has the same units as target variable:

$$RMSE = \sqrt {\frac{1}{N} \sum_{i=1}^{N} (\hat{Y_{i}} - Y_{i})^2}$$

```python
alphas = np.arange(0, 500, 5)
solvers = ['svd', 'cholesky', 'lsqr', 'sparse_cg', 'sag', 'saga']

param_grid_ridge = {
    'alpha' : alphas,
    'solver': solvers,
}

param_comb_ridge = ParameterGrid(param_grid_ridge)

# lists for storing size of errors
train_err_ridge = []
val_err_ridge = []

for params in param_comb_ridge:
    # training and evaluating each model
    model = Ridge(
        alpha=params['alpha'],
        solver=params['solver'],
        random_state=SEED
    )
    model.fit(train_data, train_target)
    train_err_ridge.append(mean_squared_error(train_target, model.predict(train_data), squared=False))
    val_err_ridge.append(mean_squared_error(val_target, model.predict(val_data), squared=False))
```

#### Results evaluation
We cannot easily visualize the model because the dimensions are too large. Same goes for the visualization of the decision boundary. We cannot plot that either because we are doing regression and not classification.


Plot the model's performance and the influence of hyperparameters:

```python
# get best parameters on validation data
best_params_ridge = param_comb_ridge[np.argmin(val_err_ridge)]

# best model
best_model_ridge = Ridge(**best_params_ridge)
best_model_ridge.fit(train_data, train_target)
result_model_ridge = best_model_ridge.predict(test_data) 

print(f"Best parameters: {best_params_ridge}\n with RMSE = {np.min(val_err_ridge)} \
and R2 score = {r2_score(test_target, result_model_ridge)}")

# display the results
plt.figure(figsize=(15,8))
for index, solver in enumerate(solvers):
    color = 'r'
    if val_err_ridge[index::len(solvers)][0] < 24: color = 'b' # solvers are overlaping, this distinguishes them
    plt.scatter(alphas, val_err_ridge[index::len(solvers)], label=solver, color=color)

plt.legend(loc="upper right")
plt.title("Solver scores comparison")
plt.xlabel("alpha")
plt.ylabel("RMSE")
plt.show()

a, b = np.polyfit(test_target, result_model_ridge, 1)
result_mean = test_target.mean()
plt.figure(figsize=(15,8))
plt.scatter(test_target, result_model_ridge)
plt.plot(test_target, test_target * a + b, 'b', label="relation between prediction and result")
ax = plt.gca()
ax.axline((0, 0), slope=1,c='g', label="ideal prediction")
ax.axhline(result_mean, c='y', label="mean of death rate")
plt.xlabel("correct result")
plt.ylabel("prediction")
plt.title("Ridge regression accuracy")
plt.xlim([70, 300])
plt.ylim([70, 300])
plt.legend(loc="upper right")
plt.show()
```

As we can se from the first graph, solvers fall into two groups. *Cholesky* and *SVD* method produce significantly smaller errors (the plots for different solvers overlap and some have identical results).

The prediction does not follow the correct results completely, however, it is still better than the mean death rate .
This indicates that the model seems to find a slight correlation between socioeconomical data and lung cancer death rate, however other factors are in play, that the model either cannot detect or are not available in the dataset, which is likelier as we would like to believe that socioeconomical factors do not influence death rates completely.

#### Other model for comparison - Random Forest
We will also train random forest and compare the results. For training will be used different approach than above. Training will be done using cross validation and therefore we won't need validation data.

```python
train_data, test_data, train_target, test_target = train_test_split(
    data.drop(columns=["TARGET_deathRate"]),
    data["TARGET_deathRate"],
    test_size=TEST_SIZE,
    random_state=SEED
)
```

We will use RMSE again.

In the model we will be tuning 3 hypermarameters:

* **n_estimators**: The number of trees in the forest.

* **min_samples_split**: The minimum number of samples required to split an internal node. For values lower than 1 it is a fraction and *ceil(min_samples_split * n_samples)* are the minimum number of samples for each split.

* **max_features**: The number of features to consider when looking for the best split.

```python
param_grid = {'n_estimators': [50, 100, 150, 200],
              'max_features' : [0.05, 0.09, 0.2, 0.36, 0.5, 0.8, 1],
              'min_samples_split' : [2, 3, 5, 8, 14, 30]
}

grid = GridSearchCV(RandomForestRegressor(random_state=SEED, n_jobs=-1), param_grid,
                    verbose=4, n_jobs=-1, return_train_score=True, scoring='neg_root_mean_squared_error')

# fitting the model for grid search 
grid.fit(train_data, train_target)
```

Same as before with ridge regression, we can't easily visualize the final model. Random forest is made from a large amount of simple trees, and while we could visualize one single tree from the forest, it wouldn't give us much insight into the model anyway because the trees can greatly differ from one another.

Next we will check how different combinations of hyperparameters influence the final prediction precision.

```python
fig,axes=plt.subplots(nrows=3,ncols=1,figsize=(7, 10))
fig.tight_layout(pad=4)

# max_features, min_samples_split
max_scores = grid_df.groupby(['param_min_samples_split', 'param_max_features']).max()
max_scores = max_scores.unstack()[['mean_test_score']]
sns.heatmap(max_scores.mean_test_score, annot=True, fmt='.5g', ax=axes[0], cmap=sns.cm.rocket_r);
axes[0].set_title('RMSE on test data (max_features, min_samples_split)', fontsize=12)

# n_estimators, min_samples_split
max_scores = grid_df.groupby(['param_min_samples_split', 'param_n_estimators']).max()
max_scores = max_scores.unstack()[['mean_test_score']]
sns.heatmap(max_scores.mean_test_score, annot=True, fmt='.5g', ax=axes[1], cmap=sns.cm.rocket_r);
axes[1].set_title('RMSE on test data (n_estimators, min_samples_split)', fontsize=12)

# n_estimators, max_features
max_scores = grid_df.groupby(['param_max_features', 'param_n_estimators']).max()
max_scores = max_scores.unstack()[['mean_test_score']]
sns.heatmap(max_scores.mean_test_score, annot=True, fmt='.5g', ax=axes[2], cmap=sns.cm.rocket_r);
axes[2].set_title('RMSE on test data (n_estimators, max_features)', fontsize=12)
```

Above are 3 heapmaps displaying RMSE in relation with a pair of hyperparameters. If we look at the first one (*max_features, min_samples_split*), it seems that best parameters are somewhere in the middle and the score gets worse with values more on the edge of the given lists.

On the second heatmap (*min_samples_split, n_estimators*), the best scoore seems to be in the top right corner. That means maximum number of estimators and minimum of *min_samples_split*.

On the third heatmap it looks like we again get better score if we choose *max_features* somewhere from the middle of our list of parameters (0.5 or 0.36). *n_estimators* doesn't play a big role in final score here.

```python
results = grid_df[["param_max_features", "param_min_samples_split",
                   "param_n_estimators", "mean_test_score"]].astype(float)

fig = px.parallel_coordinates(
    results, 
    color='mean_test_score', 
    color_continuous_scale=px.colors.diverging.RdYlGn,
)

fig.show()
```

On this graph, we can see that the conclusions we made with the heatmaps were correct. The more estimators the model had, the better results it produced. Attribute *max_features* improves score if it's somewhere in the middle, specifically 0.5 or 0.36. And the parameter *min_samples_split* doesn't seem to have much influence over the final results.
